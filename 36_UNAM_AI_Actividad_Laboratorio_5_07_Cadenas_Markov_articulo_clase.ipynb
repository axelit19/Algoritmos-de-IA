```python
# Se muestra en cantidad el conjunto de datos de entrenamiento y el conjunto de pruebas calculando el tamaño de ambos conjuntos.
print('Conjunto de entrenamiento',len(train_data))
print('Conjunto de pruebas',len(test_data))
```

```python
# Se entrena el modelo con hmm.HiddenMarkovModelTrainer al crear algoritmos para aprender parámetros HMM a partir de
# de datos de entrenamiento, mismo que incluyen tanto el aprendizaje supervisado (MLE) como el aprendizaje no supervisado.
trainer = hmm.HiddenMarkovModelTrainer()
tagger = trainer.train_supervised(train_data)
```

```python
# Se muestra el reultado de la función tagger que entrena el HMM usando el algoritmo Baum-Welch para
# maximizar la probabilidad de la secuencia de datos.
print(tagger)
```

```python
# Se muestra el resultado de los datos de prueba del modelo.
print(tagger.evaluate(test_data))
```

```python
# Se realiza el entrenamiento de palabras con la cración de un iterador "zip" agregando elementos de dos o mas variables,
# de forma que se entrene el modelo previo con los porcentajes de evaluacion anteriores para mejorar el modelo.
train_words, train_tags = zip(*chain(*train_data))
test_words, test_tags = zip(*chain(*test_data))
```

```python
# Se muestra el testeo de un set de palabras, con los niveles de error en los datos de prueba previos sugieren que el modelo se
# sobreentrenó, por lo que los datos de entrenamiento son insuficientes para cubrir los casos del conjunto de pruebas, por
# consecuente se muestra un nuevo conjunto de palabras para complementar el modelo.
len(set(test_words))
```

```python
# Mejorando el conjunto de datos de prueba de entrenamiento que el procentaje anterior mostraba como inexistente,
# se complementa con el entrenemiento de diferentes palabras para el modelo.
len(set(test_words).difference(train_words))
```

```python
# Con la funcion de PorterStemmer que realiza la extracción de la raíz de una palabra previa al entrenamiento para
# mejorar el rendimiento del modelo en 44.03%, cambiando y modificando los rangos de palabras o datos.
porter = PorterStemmer()

train_data = [ [(porter.stem(word.lower()), tag) for word, tag in sent] for sent in treebank.tagged_sents()[:3000]]
test_data = [ [(porter.stem(word.lower()), tag) for word, tag in sent] for sent in treebank.tagged_sents()[3000:]]

trainer = hmm.HiddenMarkovModelTrainer()
tagger = trainer.train_supervised(train_data)
tagger.evaluate(test_data)
```

```python
# Cambio del modelo con HiddenMarkovModelTagger para entrenar de nuevo los datos.
tagger = nltk.HiddenMarkovModelTagger.train(train_data)
```

```python
# Evaluación del nuevo modelo con las muestra del rendimiento mejorada del modelo con 83.91% de rendimiento.
tagger.evaluate(test_data)
```
